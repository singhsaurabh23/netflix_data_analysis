<h4>Individual country with the Highest No. of TV Shows.
</h4>

```python
netflix_tvshow = netflix[netflix['Category'] == 'TV Show']
country_counts = netflix_tvshow['Country'].value_counts()
```

```python
netflix_tvshow.Country.value_counts()
```

```python

# Plotting the bar chart
plt.figure(figsize=(12, 6))
country_counts.head(10).plot(kind='bar', color='skyblue')  # Show top 10 countries
plt.title('Top 10 Countries with the Most Netflix TV Shows')
plt.xlabel('Country')
plt.ylabel('Number of TV Shows')
plt.xticks(rotation=45)
plt.show()
```

<h4>Individual country with the Highest No. of Movies.</h4>

```python
netflix_movie = netflix[netflix['Category'] == 'Movie']
country_counts2 = netflix_movie['Country'].value_counts()
plt.figure(figsize=(12, 6))
country_counts2.head(10).plot(kind='bar', color='skyblue')  # Show top 10 countries
plt.title('Top 10 Countries with the Most Netflix Movies')
plt.xlabel('Country')
plt.ylabel('Number of Movies')
plt.xticks(rotation=45)
plt.show()
```

<h4> Trend of movies and tv shows released per year </h4>

```python
# Convert Release_Date to datetime
netflix['Date_N'] = pd.to_datetime(netflix['Release_Date'], errors='coerce')

# Extract Year from Release_Date
netflix['Year'] = netflix['Date_N'].dt.year

# Grouping by Year and Category, and counting the number of movies and TV shows
trend_data = netflix.groupby(['Year', 'Category']).size().unstack(fill_value=0)

# Plotting the line chart
plt.figure(figsize=(12, 6))
for column in trend_data.columns:
    plt.plot(trend_data.index, trend_data[column], marker='o', label=column)

plt.title('Number of Movies and TV Shows Released Per Year')
plt.xlabel('Year')
plt.ylabel('Number of Titles Released')
plt.legend(title='Category')
plt.grid()
plt.xticks(trend_data.index, rotation=45)

plt.show()
```
